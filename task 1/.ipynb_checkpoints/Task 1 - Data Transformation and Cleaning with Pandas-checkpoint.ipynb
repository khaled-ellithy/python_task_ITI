# Read dataset 

````python
import pandas as pd
import numpy as np

Energy = pd.read_excel('Energy Indicators.xls',skiprows=17 ,skipfooter=38,
          usecols=[2,3,4,5])
Energy.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
Energy
````

````python
Energy['Energy Supply']=[i if str(i).isdigit() else np.nan for i in Energy['Energy Supply'] ]
Energy
````

````python
Energy['Energy Supply']=Energy['Energy Supply']*1000000
Energy
````

# change names of required countries

````python
Energy['Country']=np.where(Energy['Country']=='Republic of Korea','South Korea',Energy['Country'])
Energy['Country']=np.where(Energy['Country']=='United States of America20','United States',Energy['Country'])
Energy['Country']=np.where(Energy['Country']=='United Kingdom of Great Britain and Northern Ireland19','United Kingdom',Energy['Country'])
Energy['Country']=np.where(Energy['Country']=='China, Hong Kong Special Administrative Region3','Hong Kong',Energy['Country'])
````

# clean name of country after removing "(", ","

````python
def keep_country(text):
    if '(' in text:
        country=text.split('(')
    else:   
        country=text.split(',')
    clean_text = ''.join([item for item in country[0] if not item.isdigit()])
    return clean_text.strip()
````

````python
Energy['Country'] = Energy['Country'].apply(keep_country)
````

# take needed columns from GDP and join the three dataset using reduce

````python
from functools import reduce

GDP=GDP.loc[:,'2006':'Country']
df_merge=[GDP,ScimEn,Energy]
total = reduce(lambda left,right: pd.merge(left,right,on='Country'), df_merge)
total.drop_duplicates(subset ="Country",inplace=True)
total.set_index('Country',inplace=True)
total
````

````python
total.sort_values(by=['Rank'],ascending=True,inplace=True)
top_total=total.iloc[0:15,:].copy()
top_total
````

Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

ContinentDict  = {'China':'Asia',

*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

df=top_total.loc[:,'2006':'2015'].copy()
count_of_years=len(df.columns)
df['continent']=ContinentDict.values()
df['Country']=df.index
df = df.melt(id_vars=['Country','continent'])
df
````

````python
df=df.groupby('continent').agg([np.size,np.sum, np.mean,np.std])
df['value']['size']=df['value']['size']/count_of_years
df
````
